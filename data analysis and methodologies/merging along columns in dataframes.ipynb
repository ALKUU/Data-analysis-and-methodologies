```python
import pandas as pd
```

```python
excelfile=pd.ExcelFile('Book 1.xlsx')
df_continent=excelfile.parse('Sheet1')
df_country=excelfile.parse('Sheet2')
print(df_continent)
print(df_country)
```

```python
pd.merge(df_continent,df_country,on='conti.ID',how='left')
```

```python
pd.merge(df_continent,df_country,on=['conti.ID','var'],how='left')
```

```python
#using suffixes
pd.merge(df_continent,df_country,on='conti.ID',how='left',suffixes=('_continent','_country'))
```
